```python
import librosa
import numpy as np
```

```python
SAMPLING_RATE=16000
DURATION=10
N_FFT = 1024
WIN_LENGTH = 1024
HOP_LENGTH = 512
```

```python
music_audio = np.load('/content/drive/MyDrive/data/music_audio.npy')
speech_audio = np.load('/content/drive/MyDrive/data/speech_audio.npy')
```

```python
dur = 3
l = dur*SAMPLING_RATE
samples = speech_audio.shape[0]/l
samples = int(samples)

music_audio = music_audio[:samples*l]
speech_audio = speech_audio[:samples*l]

music_audio = music_audio.reshape(-1, l)
speech_audio = speech_audio.reshape(-1, l)
```

```python
print(music_audio.shape)
```

```python
final_aud = np.array([])
n = music_audio.shape[0]
events = []
cur_time = 0
a = 0
b = 0
while(a<n or b<n):
  if np.random.randint(2) == 1 and a<n:
    final_aud = np.append(final_aud, music_audio[a])
    events.append([cur_time, cur_time+3, 'music'])
    cur_time = cur_time + 3
    a = a+1
  elif np.random.randint(2) == 0 and b<n:
    final_aud = np.append(final_aud, speech_audio[b])
    events.append([cur_time, cur_time+3, 'speech'])
    cur_time = cur_time + 3
    b = b+1
  
  sil_len = 1 + np.random.rand()*2
  sil_frames = int(sil_len*SAMPLING_RATE)
  S = np.zeros(sil_frames)
  final_aud = np.append(final_aud, S)
  cur_time = cur_time + sil_len
print(final_aud.shape)
```

```python
print(events)
```

```python
modified_events = []
for event in events:
  if event[1]%10 < event[0]%10:
    diff1 = 9.9999999-(event[0]%10)
    diff2 = 10.0000001-(event[0]%10)
    modified_events.append([event[0], event[0]+diff1, event[2]])
    modified_events.append([event[0]+diff2, event[1], event[2]])
  else:
    modified_events.append([event[0], event[1], event[2]])

print(modified_events)
```

```python
l = final_aud.shape[0]/160000
l = int(l)
l = l*160000
final_aud_trimmed = final_aud[:l]
X_audios = np.reshape(final_aud_trimmed, (-1, 160000))
print(X_audios.shape)
```

```python
id = 0
event_list = []
t = 10
for event in modified_events:
  if(event[0]>t):
    t = t+10
    id = id + 1
  event_list.append([id, event[0], event[1], event[2]])
```

```python
print(event_list)
```
